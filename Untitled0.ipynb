```python
# DFS using Recursion

graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": [],
    "E": [],
    "F": []
}

visited = []

def DFS(node):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)

        for i in graph[node]:
            DFS(i)

# Main Program
print("DFS Traversal is:")
DFS("A")
```
